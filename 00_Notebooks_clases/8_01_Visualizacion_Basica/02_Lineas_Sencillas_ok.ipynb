### Anatoma de una Figura y un Axe 


Para todos los grficos de Matplotlib, comenzamos creando una figura y unos ejes (objeto tipo *axes*, ojo eje en ingls se dice *axis*). [recuerda que una figura es un conjunto de una o ms grficas que llamamos ejes o mejor axes]
En su forma ms simple, una figura y unos axes (*ejes*) se pueden crear de la siguiente manera:

```python
fig = plt.figure()
ax = plt.axes()
```

En Matplotlib, la *figura* (una instancia de la clase ``plt.Figure``) puede ser considerada como un solo **contenedor que contiene todos los objetos que representan ejes, grficos, texto y etiquetas.**  

Los *ejes* (una instancia de la clase ``plt.Axes``) es lo que vemos arriba: un cuadro delimitador con **marcas y etiquetas, que eventualmente contendr los elementos del grfico que conforman nuestra visualizacin.**  

A lo largo del bootcamp, y en general lo vas a ver as fuera de l, comnmente usaremos el nombre de la variable ``fig`` para referirnos a una instancia de figura, y ``ax`` para referirnos a una instancia de ejes o grupo de instancias de ejes.

Como trataremos en sesiones posteriores en realidad lo que ves son los elementos de un axe (eje) y una figura es un conjunto de axes

### Primeros grficos (axes)


Una vez que hemos creado unos ejes, podemos usar la funcin ``ax.plot`` para trazar algunos datos. Comencemos con la funcin seno:

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x,np.sin(x));
```

Alternativamente, podemos usar la interfaz de pylab y dejar que la figura y los ejes se creen automticamente en segundo plano

```python
plt.plot(x, np.sin(x)); # Si solo es para pintarla y ya esta. 
```

Si queremos crear una sola figura con mltiples lneas, simplemente podemos llamar a la funcin `plot` varias veces:

```python
plt.style.use("tableau-colorblind10")
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x));
```


Eso es todo lo que hay que hacer para trazar funciones simples en Matplotlib. Recuerda que en realidad lo que hace es pintar puntos asociados y unirlos.


```python
y = np.random.randint(0,12,len(x))
ax.plot(x,y)
display(fig)
```

### Ajustando el Grfico: Colores y Estilos de Lnea

El primer ajuste que podras desear hacer a un grfico es controlar los colores y estilos de lnea.
La funcin ``plt.plot()`` toma argumentos adicionales que se pueden usar para especificar estos.
Para ajustar el color, puedes usar la palabra clave ``color``, que acepta un argumento de tipo string que representa virtualmente cualquier color imaginable.
El color se puede especificar de varias maneras:

```python
# specify color by name
plt.plot(x,np.sin(x - 0), color = 'black')
```

```python
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

Si no se especifica ningn color, Matplotlib automticamente recorrer un conjunto de colores predeterminados para mltiples lneas.

Del mismo modo, el estilo de lnea se puede ajustar utilizando la palabra clave ``linestyle``:

```python
plt.plot(x,x+1,linestyle = '--')
```

```python
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

Si quisieras ser extremadamente conciso, estos cdigos de ``linestyle`` y ``color`` pueden combinarse en un solo argumento no clave para la funcin ``plt.plot()``:

```python
plt.plot(x,x,"g")
```

```python
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

Estos cdigos de color de un solo carcter reflejan las abreviaturas estndar en los sistemas de color RGB (Rojo/Verde/Azul) y CMYK (Cian/Magenta/Amarillo/negro), comnmente utilizados para grficos de color digital.

Hay muchos otros argumentos clave que se pueden utilizar para ajustar finamente la apariencia del grfico; para ms detalles, como siempre, ir a la [documentacin de plt.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot)


Usarlo con objetos axe es "directo" (por resumir slo en la versin concentrada):


```python
fig = plt.Figure()
ax = plt.axes()
ax.plot(x, x + 0, '-g')  # solid green
ax.plot(x, x + 1, '--c') # dashed cyan
ax.plot(x, x + 2, '-.k') # dashdot black
ax.plot(x, x + 3, ':r');  # dotted red

```


### Etiquetado de grficos

#### Titulo, etiqueta x, etiqueta y

Como ltima parte de esta sesin, echaremos un breve vistazo al etiquetado de grficos: ttulos, etiquetas de ejes y leyendas simples.

Los ttulos y las etiquetas de ejes son las etiquetas ms simples de este tipo:

```python
plt.plot(x, np.sin(x))
plt.title("La funcin seno")
plt.xlabel("Valores")
plt.ylabel("y");
```

```python
fig = plt.figure()
ax = plt.axes()
ax.set_xlim(0,20)
ax.set_ylim(-1,2)
ax.plot(x,np.sin(x))
ax.set_title("Otro Grafico")
ax.set_xlabel("x")
ax.set_ylabel("y")
```


**La posicin, tamao y estilo de estas etiquetas se pueden ajustar utilizando argumentos opcionales en la funcin.**
Para obtener ms informacin, consulte la documentacin de Matplotlib y las cadenas de documentacin de cada una de estas funciones.


#### Leyenda


**Matplotlib tiene una forma incorporada de crear rpidamente una leyenda**, cuando hay varias figuras:

```python
plt.plot(x,np.sin(x), label = "Seno Ejemplo")
plt.plot(x,np.cos(x), label = "coseno")
plt.legend()
```

```python
fig = plt.figure()
ax = plt.axes()
ax.set_xlim(0,20)
ax.plot(x, np.sin(x), label = "Sen")
ax.plot(x, np.cos(x), label = "Cos", ls = "-.", color = "g")
ax.legend()
ax.set_title("Pruebas y no funciona")
ax.set_xlabel("Valores axe")
ax.set_ylabel("Valores y en axe");
```
