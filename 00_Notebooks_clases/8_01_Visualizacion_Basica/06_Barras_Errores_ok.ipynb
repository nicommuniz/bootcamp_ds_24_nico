### Graficos de Barras y Count plots

Los grficos de barras se emplean para comparar datos de forma ms general y cuando se emplean para comparar frecuencias tambin reciben el nombre de Count plots. La sintaxis del grfico de barras en matplotlib es muy sencilla:

```python

# Datos de ejemplo
categorias = ['Categora A', 'Categora B', 'Categora C']
valores = [23, 45, 56]

# Crear grfico de barras
plt.bar(categorias, valores);

```

Pero si queremos obtener un grfico de frencuencias le tenemos que dar estas calculadas, veamos como sera el grfico de frecuencias absolutas y de frecuencias relativas para la columna "who" del titanic, empleando ya algunos de nuestros aprendizajes:

```python
# frecuencias absolutas
categorias = df_titanic.who.value_counts().index.to_list()
frecuencias = df_titanic.who.value_counts().to_list()
frec_relativas = (df_titanic.who.value_counts(normalize = True) * 100).to_list()
fig,ax = plt.subplots(1,2, figsize = (4.5,3))
fig.subplots_adjust(wspace = 0.5)
ax[0].bar(categorias, frecuencias)
ax[0].set_title("Frecuencias Absolutas")
ax[0].set_xlabel("Valores Who")
ax[0].set_ylabel("Num. personas")
ax[1].bar(categorias, frec_relativas, color = ["brown","red","steelblue"])
ax[1].set_title("Frecuencias Relativas")
ax[1].set_xlabel("Valores Who")
ax[1].set_ylabel("% del total");
```


### Visualizando Errores


En general, si no vas a hacer un anlisis de datos de mediciones no vas a necesitar este tipo de grficas, pero son muy comunes en mbitos cientficos y para representar medidas que por las condiciones de medicin o de los aparatos de medida quieren presentar el posible rango de error que se haya podido cometer al tomarlas. Por ejemplo si necesitamos mucha precisin en la medida de la velocidad de un objeto en un experimento de fsica y nuestro aparato de medida dice 5 pero tambin que comete un error de 0.05, lo normal es representar ese 5 como un punto y una barra, denominada barra de error, centrada en la medida y de extremos el resultado de restar y sumar el error (en este caso entre 4.95 y 5.05, resultado de restar 0.05 y sumar 0.05)



En la visualizacin de datos y resultados, mostrar estos errores de manera efectiva puede hacer que una grfica transmita mucha ms informacin completa.



Cmo se muestran barras de error bsico en matplotlib:

```python
x = np.linspace(0, 10, 50)
dy = 0.8 # el error de nuestro aparato de medida
y = np.sin(x) + dy * np.random.randn(50) # una funcin inventada con ese error (aleatorio por medida)

plt.errorbar(x, y, yerr=dy, fmt='.g');
```

Adems de estas opciones bsicas, la funcin ``errorbar`` tiene muchas opciones para ajustar finamente las salidas.
Usando estas opciones adicionales, puedes personalizar fcilmente la esttica de tu grfico de barras de error.
A menudo encuentro til, especialmente en grficos concurridos, hacer que las barras de error sean ms claras que los propios puntos:

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=2, capsize=3);
```

Existen muchas ms opciones, que puedes encontrar en la documentacin de `plt.errorbar`, pero por lo menos tienes la forma ms sencilla de mostrar errores si esa es tu necesidad.
