### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Figura 1

#Obtengo las ciudades mas pobladas con ciudad y cantidad de poblacion
ciudades_mas_pobladas = df_cali.loc[df_cali.population_total.nlargest(5).index, 
                                    ["city","population_total"]]
ciudades_mas_pobladas

#Para obtener las ciudades en una lista
def separa_nombre(nombre):
    nuevo = nombre[0]
    for letra in nombre[1:]:
        if letra == letra.upper():
            nuevo += " "
        nuevo += letra
    return nuevo


ciudades = [separa_nombre(ciudad) for ciudad in ciudades_mas_pobladas["city"]]
ciudades

# OTRA FORMA DE OBTENER LAS CIUDADES SIN EL FOR NI LA FUNCION
# ciudades = ciudades_mas_pobladas["city"].to_list()


# Los colores en el df
colores = ["blue","gray","red","olive","brown"]

#Tamao
plt.figure(figsize = (10,5))

# Tipo y especificaciones
plt.bar(x= ciudades, height= ciudades_mas_pobladas["population_total"]/1000, color = colores, alpha = 0.5)
plt.ylim([0,5000])
plt.ylabel("Habitantes (miles)")
plt.xlabel("Ciudades");

# Para agregarle los valores 
for indice, ciudad in enumerate(ciudades_mas_pobladas.index):
    poblacion = ciudades_mas_pobladas.loc[ciudad,"population_total"]/1000
    plt.text(ciudades[indice], poblacion, f"{poblacion}", ha = "center")

plt.title("Ciudades ms pobladas de California");

```

```python
# Figura 2

plt.hist(df_cali["area_total_km2"], color = "steelblue", alpha = 0.7, bins = 50,label = "Hist Area");
plt.xlabel("Superficie ($km^2$)", size = 18);
plt.ylabel("Nmero de Ciudades", size = 18);
plt.legend();
```

```python
# FIgura 3

plt.scatter(df_cali["area_water_km2"], df_cali["population_total"]/1000, alpha = 0.5, color = "green")
plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacin (miles habitantes)")
plt.title("Poblacin por kilometro cuadrado de agua")

```

```python
# Figura 4 
plt.boxplot(df_cali["elevation_m"].dropna(), vert = False, boxprops = dict(facecolor = "red"),
               medianprops = dict(color= "black"), patch_artist= True )
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
```

```python
fig,axs = plt.subplots(2,2, figsize = (20,20))
axs = axs.flatten()

# 1
primera = axs[0]
ciudades_mas_pobladas = df_cali.loc[df_cali.population_total.nlargest(5).index, ["city","population_total"]]
ciudades = [separa_nombre(ciudad) for ciudad in ciudades_mas_pobladas["city"].to_list()]
colores = ["blue","gray","red","olive","brown"]

primera.bar(x= ciudades, height= ciudades_mas_pobladas["population_total"]/1000, color = colores, alpha = 0.5)
primera.set_ylim([0,5000])
primera.set_ylabel("Habitantes (miles)")
primera.set_xlabel("Ciudades")

ciudades_mas_pobladas.set_index("city", inplace = True)
for indice, ciudad in enumerate(ciudades_mas_pobladas.index):
    poblacion = ciudades_mas_pobladas.loc[ciudad,"population_total"]/1000
    primera.text(ciudades[indice], poblacion , f"{poblacion}", ha = "center")
primera.set_title("Ciudades ms pobladas de California")

# 2
segunda = axs[1]
segunda.hist(df_cali["area_total_km2"], color = "steelblue", alpha = 0.7, bins = 50,label = "Hist Area");
segunda.set_xlabel("Superficie ($km^2$)", size = 18);
segunda.set_ylabel("Nmero de Ciudades", size = 18);
segunda.legend();

# 3

tercera = axs[2]
tercera.scatter(df_cali["area_water_km2"], df_cali["population_total"]/1000, alpha = 0.5, color = "green")
tercera.set_xlabel("rea de agua en la ciudad ($km^2$)")
tercera.set_ylabel("Poblacin (miles habitantes)")
tercera.set_title("Poblacin por kilometro cuadrado de agua")

# 4
cuarta = axs[3]
cuarta.boxplot(df_cali["elevation_m"].dropna(), vert = False, boxprops = dict(facecolor = "red"),
               medianprops = dict(color= "black"), patch_artist= True)
cuarta.set_xlabel("Altitud (m)")
cuarta.set_title("Diagrama de caja de la Altitud de las ciudades de California"); 
```
